## Uvoz podatkov s pomočjno knjižnice *pandas*

```python
# Uvoz knjižnic za potrebe podatkovne analitike
import pandas as pd
```

```python
# Uvoz podatkov v obliki csv s pomočjo knjižnice pandas
df = pd.read_csv("https://raw.githubusercontent.com/urosgodnov/datasets/refs/heads/master/50_Startups.csv")
```

## Pregled in urejanje uvoženih podatkov

```python
# Funkcija po stolpcih pregleda, koliko je v tabeli manjkajočih funkcij
df.isna().sum()
```

Sama tabela je brez manjkajočih funkcij, zato
je primerna za nadaljno analizo. Če bi v katerem stolpcu manjkalo več kot 20% podatkov, bi bil stolpec neuporaben za nadaljno podatkovno analitiko.

```python
# Čiščenje podvojenih in manjkajočih vrednosti (sicer jih ni, ampak za obnovitev znanja)
df.dropna()
df.drop_duplicates()
```

```python
# Odstranitev vrstic, kjer je R&D Spend in Marketing Spend enak 0
df = df[(df["R&D Spend"] != 0) | (df["Marketing Spend"] != 0)] # to je predlagal Gemini, deluje na logiki, da odstrani vrstice z ničlami v definiranih stolpcih
df # izris nove tabele
```

Vrstice, kjer sta R&D Spend in Marketing Spend hkrati enaka 0 sem odstranila. Za nadaljno analizo sem Profit določila kot odvisno spremenljivko. Vrsitce sem odstranila, ker sem mnenja, da ne pripomorejo moje nadaljni raziskavi.

```python
# Pregled kako popolna je tabela
df.info()
```

Iz izpisa razberem, da tabela preurejena vsebuje 49 vrstic in 5 stolpcev, 4 stoplci imajo numerične podatke (decimalna oblika), en stolpec ima opisne podatke.
Tabela zavzema 2,3 KB prostora. Vsi stolpci imajo popolne podatke, torej ni manjkajočih vrednosti.

## Uvedba nove opisne spremenljivke

```python
# Uvedba opisne spremenljivke za stolpec Administration
df["Administration_cost_objectiv"] = df["Administration"].apply(lambda x: "high" if x > 100000 else "low")

df # dodano zato, da se izpiše tabela
```

## Histogrami in linearna regresija

```python
# Razni histogrami
df.hist()
```

```python
# Povezanost višine dobička z vložkom v marketing

import plotly.express as px # Vploklic nove knjižnice, za hitro analizo podatkov
fig = px.scatter(df, x="Marketing Spend", y="Profit", trendline="ols") # Uvedba nove figure, definiranje odvisne in neodvisne spremenljivke
fig.show() # Prikaži figuro
```

## Profile report

Koda za Profile Report ni napisana v optimalni obliki, ker ni delovala. Trenutna koda je skupek kode iz predavanja in pa kode iz Chat GPT-ja. Nesmiselno se mi zdi podnovno uvažanje podatkov, ampak po 7-ih neuspelih poskusih sem sprejela poraz in uporabila to kodo.

```python

!pip install ydata-profiling # Namestitev knjižnice za profile report

# 2. Uvozi potrebne knjižnice
from ydata_profiling import ProfileReport # Uvoz knjižnice

# 3. Naloži svoj DataFrame (lahko nadomestiš s svojim csv)
df2 = pd.read_csv("https://raw.githubusercontent.com/urosgodnov/datasets/refs/heads/master/50_Startups.csv") # Ponoven uvoz dataframe

target_variable = 'Profit' # Definiranje odvisne spremenljivke

profile = ProfileReport (df2, title='Profile Report') # Definiranje in poimenovanje poročila
profile # Prikaz poročila
```

Poročilo nakazuje, da dataframe mogoče ni najbolj primeren za moj analitični primer, predvsem ker gre za dokaj majhno bazo podatkov, ki je bila že v prvotni obliki skoraj popolna (brez podvojenih, manjkajočih vrednosti), kljub temu pa se še vedno da izvajati smiselne analize.

Kot že omenjeno sem za odvisno spremenljivko izbrala *Profit*, zanimalo me je predvsem, kako velikost različnih skupin stroškov (Marketing, R&D) vpliva na končni dobiček.

Iz reporta lahko razberem, da stroški Marketinga in R&D linearno vplivajo na rast dobička, več denarja kot podjetje zapravi za te dve kategoriji, večji je njihov dobiček. Pozitivna korelacija med R&D in dobičkom je močnejša, kot pozitivna korelacija med Marketingom in dobičkom.

```python
# Podrobnejši pregled koreacije med profitom in stroški administracije
import statsmodels.formula.api as smf

model = smf.ols(formula='Profit ~ Administration', data=df).fit()
print(model.summary())
```
